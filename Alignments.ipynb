# Import of Sequences

```python
import os 
os.chdir(r'C:\Users\Teresa\OneDrive - University Of Cambridge\St Johns Cambridge\1-Part II Biochemistry\Part II Project\Files\Randomer sequencing')

import sys
from Bio.Seq import Seq
# def to read fasta files
def readFasta(fastaFile):
    fh = open(fastaFile, 'r')
    for line in fh:
        if line[0] == '>':
            header = line.rstrip()[1:]
            if sys.version_info[0] < 3:
                seq = fh.next().rstrip()
            else:
                seq = fh.readline().rstrip()
        yield [header, seq]
    fh.close()

dict_rounds = {}


rounds_names = ['RZ5N','F201N', 'F205N', 'N17L', 'N73L', 'N73H', 'FN17L', 'FN73L','FN73H'] #paper

count = {}
for i in rounds_names:
    dict_rounds[i]={}
    count[i]=0
    for line in readFasta(i+'.fasta'): # looks through each line of each file
        count[i]+=1
        if  line[1] not in dict_rounds[i]: #adds to dict if seq not there
            dict_rounds[i][line[1]] = 1
            
        else:
            dict_rounds[i][line[1]] +=1            
```

# Filter

```python
dict_filt3_range_rounds = {} #only take sequences that are multiples of 3 within right range 
#except pools with PNK are not multiple of 3 but right range and above 8 nt
dict_filt3_rest={} #all sequences that are not multiple of 3

def is_integer_num(n):
    if isinstance(n, int):
        return True
    if isinstance(n, float):
        return n.is_integer()
    return False

count_range = {} #new counting dictionary to include different categories
length_count ={} #counts number of sequences in pool at length all in range and filtered 3


count_categories=['total','greater5len','filtered3_greater5len','in_range','in_range_filtered3', 
                  'fraction_greater5len_total',
                  'fraction_filtered3_greater5len_total',
                  'fraction_in_range_total',
                  'fraction_in_range_filtered3_total']

for x in rounds_names:
    count_range[x] = dict.fromkeys([count_categories[y] for y in range(len(count_categories))],0)
    
for i in rounds_names:
    dict_filt3_range_rounds[i]={}
    length_count[i]={}
    dict_filt3_rest[i]={}
    for seq in dict_rounds[i]:
        count_range[i]['total']+= dict_rounds[i][seq]
        
        if len(seq)>5: #sequences with length 6 or more
            count_range[i]['greater5len']+=dict_rounds[i][seq]
            
            if  seq not in dict_filt3_range_rounds[i]:
                if is_integer_num(len(seq)/3): #sequences with length multiple of 3
                    count_range[i]['filtered3_greater5len']+=dict_rounds[i][seq]
                    
                    if i == 'N17L' or i =='N73L' or i =='FN17L' or i =='FN73L': #range 12-30 nt plus minus 6 (below 6 already filtered before)
                        if len(seq) <36:

                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq]
                            count_range[i]['in_range_filtered3'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]

                    elif i == 'N73H' or i =='FN73H': #range 33-150 nt plus minus 6 
                        #--> adjusted to lower end as most reads seem to be shorter in pool
                        if len(seq) > 21 and len(seq) < 156:
                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq] 
                            count_range[i]['in_range_filtered3'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]

                    elif i == 'FN847':
                        if len(seq) > 15 and len(seq) < 33: #around 27 nt plus minus 6 
                            #adjusted to lower end as most reads here seem to be shorter (see plot)
                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq]
                            count_range[i]['in_range_filtered3'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]

                    elif i == 'RZ1N' or i == 'RZ5N' or i== 'DG1N' or i == 'DG5N' or i == 'F201N' or i =='F205N':
                        if len(seq) < 27: #6-21 nt plus minus 6
                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq]
                            count_range[i]['in_range_filtered3'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]

                    elif i == 'N145' or i == 'FN145':
                        if len(seq) < 51: #9-45 nt plus minus 6
                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq]
                            count_range[i]['in_range_filtered3'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]
                                
                    elif i == 'N73A' or i == 'FN73A': #range 12-150 nt plus minus 6
                        if len(seq) <156:
                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq] 
                            count_range[i]['in_range_filtered3'] += dict_rounds[i][seq]
                            count_range[i]['in_range'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]
                else: #takes additional sequences for these pools without requiring multiple of 3
                    if i == 'N73A' or i == 'FN73A': #range 12-150 nt plus minus 6
                        if len(seq) <156:
                            dict_filt3_range_rounds[i][seq]=dict_rounds[i][seq] 
                            count_range[i]['in_range'] += dict_rounds[i][seq]
                            
                            if len(seq) not in length_count[i]:
                                length_count[i][len(seq)]=dict_rounds[i][seq]
                            else:
                                length_count[i][len(seq)]+=dict_rounds[i][seq]
                    if not is_integer_num(len(seq)/3):
                        
                        if i == 'N17L' or i =='N73L' or i =='FN17L' or i =='FN73L': #range 12-30 nt plus minus 6 (below 6 already filtered before)
                            if len(seq) <36:

                                dict_filt3_rest[i][seq]=dict_rounds[i][seq]

                                

                        elif i == 'N73H' or i =='FN73H': #range 33-150 nt plus minus 6 
                            #--> adjusted to lower end as most reads seem to be shorter in pool
                            if len(seq) > 21 and len(seq) < 156:
                                dict_filt3_rest[i][seq]=dict_rounds[i][seq] 

                                
                        elif i == 'RZ1N' or i == 'RZ5N' or i== 'DG1N' or i == 'DG5N' or i == 'F201N' or i =='F205N':
                            if len(seq) < 27: #6-21 nt plus minus 6
                                dict_filt3_rest[i][seq]=dict_rounds[i][seq]

                            
                                                        
for i in rounds_names:
    count_range[i]['fraction_greater5len_total'] = count_range[i]['greater5len']/count_range[i]['total']
    count_range[i]['fraction_filtered3_greater5len_total']=count_range[i]['filtered3_greater5len']/count_range[i]['total']
    count_range[i]['fraction_in_range_total'] = count_range[i]['in_range']/count_range[i]['total']
    count_range[i]['fraction_in_range_filtered3_total']=count_range[i]['in_range_filtered3']/count_range[i]['total']
    
```

# Alignments

## Alignment all

```python
#takes very long to run
type5 = 'GGATCTTCTCGATCTAACAAAAAAGACAAATCTGCCACAAAGCTTGAGAGCATCTTCGGATGCAGAGGCGGCAGCCTTCGGTGGCGCGATAGCGCCAACGTTCTCAACTATGACACGCAAAACGCGTGCTCCGTTGAATGGAGTTTATCATG'
type5_reversecomp = 'CATGATAAACTCCATTCAACGGAGCACGCGTTTTGCGTGTCATAGTTGAGAACGTTGGCGCTATCGCGCCACCGAAGGCTGCCGCCTCTGCATCCGAAGATGCTCTCAAGCTTTGTGGCAGATTTGTCTTTTTTGTTAGATCGAGAAGATCC'
type1 = 'GACCAATCTGCCCTCAGAGCCCGAGAACATCTTCGGATGCAGAGGAGGCAGGCTTCGGTGGCGCGATAGCGCCAACGTCCTCAACCTCCAATGCATCCCACCACATGATGAGCCTGAAGAGCCTTGGTTTTTTTG'
type1_reversecomp = 'CAAAAAAACCAAGGCTCTTCAGGCTCATCATGTGGTGGGATGCATTGGAGGTTGAGGACGTTGGCGCTATCGCGCCACCGAAGCCTGCCTCCTCTGCATCCGAAGATGTTCTCGGGCTCTGAGGGCAGATTGGTC'

ribozyme = [type5, type5_reversecomp, type1, type1_reversecomp]
ribozyme_names = ['type5', 'type5_reversecomp', 'type1', 'type1_reversecomp']


dict_alignment = {} # structure: pool --> ribozyme --> position --> score:/sequence:
dict_not_ribozyme = {}
dict_ribozyme={}
dict_grey_area={} #not ribozyme and not not ribozyme --> in between --> not assigned
alignment_count={}
dict_alignment_noplusandminus = {}


from Bio import pairwise2


pool_selection = ["RZ5N","F201N","F205N","N17L","N73L","N73H", "FN17L","FN73L","FN73H"]



for i in pool_selection:
    count = 0 # counts total number of sequences in pool
    dict_alignment[i]={}
    help_dict_align={} #temperorary dictionary 
    alignment_start={} #temporary dictionary
    alignment_end={} #temporary dictionary
    
    #dictionaries collecting output sequences
    dict_ribozyme[i]={}
    dict_grey_area[i]={}
    dict_not_ribozyme[i]={}
    
    #collects counts of sequences in each pool
    alignment_count[i]={}
    alignment_count[i]['aligned']=0
    alignment_count[i]['grey_area']=0
    alignment_count[i]['not_ribozyme']=0
    alignment_count[i]['plus_aligned']=0
    alignment_count[i]['minus_aligned']=0
    alignment_count[i]['plus_and_minus']=0
    alignment_count[i]['fraction_plus_aligned']=0
    alignment_count[i]['fraction_minus_aligned'] = 0
    alignment_count[i]['fraction_plus_and_minus'] =0
    
   
    dict_alignment_noplusandminus[i] = {}
    
    for r in range(len(ribozyme)):
        r_name = ribozyme_names[r]
        dict_alignment[i][r_name]={} #add subdictionary
        dict_alignment_noplusandminus[i][r_name]={}

        r_len = len(ribozyme[r]) #length of ribozyme
        positions_list = []
        for z in range(r_len):
            positions_list.append(z) #list of all positions ribozyme
                
        for p1 in positions_list: #prepare dictionary with all positions --> makes sure only record positions that exist as 
            #type 1 shorter than t5
            dict_alignment[i][r_name][p1]={}
            dict_alignment[i][r_name][p1]['score']=0
            dict_alignment[i][r_name][p1]['sequence']=ribozyme[r][p1]
            dict_alignment_noplusandminus[i][r_name][p1]={}
            dict_alignment_noplusandminus[i][r_name][p1]['score']=0
            dict_alignment_noplusandminus[i][r_name][p1]['sequence']=ribozyme[r][p1]
        
        dict_ribozyme[i][r_name]={}
        
    
    for seq in dict_filt3_range_rounds[i]: #go through each file of filtered dictionary
        #can insert length limitation here if only want to include length of sequences not filtered out previously
        #change cutoff values here:
        seq_len = len(seq)
        cutoff_ribozyme = 0.66+2/seq_len
        cutoff_not_ribozyme = 0.6
        
        count += dict_filt3_range_rounds[i][seq]
        
        aligned = 0
        not_ribozyme = 0
        alignment_start[seq]={}
        alignment_end[seq]={}
        help_dict_align[seq]={}
        
        for r1 in range(len(ribozyme)):
            alignment = pairwise2.align.localms(ribozyme[r1],seq,1,0,-1000,-1000) #1 for match, 0 for mismatch, -1000 for gap
            alignment_score = alignment[0][2]
            alignment_score_normalized = alignment_score/seq_len
                
            if alignment_score_normalized <cutoff_not_ribozyme:
                not_ribozyme +=1
            
            if alignment_score_normalized > cutoff_ribozyme: #cut off adjusted for length --> determined through looking at alignment distributions for different pools
                aligned+=1
                help_dict_align[seq][ribozyme_names[r1]] = dict_filt3_range_rounds[i][seq]
                dict_ribozyme[i][ribozyme_names[r1]][seq]=dict_filt3_range_rounds[i][seq]
                alignment_start[seq][ribozyme_names[r1]]=alignment[0][3] #retrieves start
                alignment_end[seq][ribozyme_names[r1]]=alignment[0][4] #retrieves end
                
        if aligned == 0:
            if not_ribozyme == 4:
                alignment_count[i]['not_ribozyme']+= dict_filt3_range_rounds[i][seq]  
                dict_not_ribozyme[i][seq]=dict_filt3_range_rounds[i][seq]
            else:
                alignment_count[i]['grey_area']+=dict_filt3_range_rounds[i][seq]
                dict_grey_area[i][seq]=dict_filt3_range_rounds[i][seq]
        else:
            alignment_count[i]['aligned']+=dict_filt3_range_rounds[i][seq]
            
    for seq in help_dict_align: #write scores into alignment dictionary and triplet match
        plus_strand = 0
        minus_strand =0
        
        for r2 in range(len(ribozyme)):
            if ribozyme_names[r2] in help_dict_align[seq]:
                if (ribozyme_names[r2] == 'type5') or (ribozyme_names[r2] == 'type1'):
                    plus_strand +=1
                    
                else:
                    minus_strand +=1
                    
                my_range = alignment_end[seq][ribozyme_names[r2]] - alignment_start[seq][ribozyme_names[r2]]
                for y in range(my_range):
                    position = alignment_start[seq][ribozyme_names[r2]]+y #getting position on ribozyme to put into dictionary for each nucleotide in sequence
                    
                    s = dict_filt3_range_rounds[i][seq]/(count*len(help_dict_align[seq]))
                    dict_alignment[i][ribozyme_names[r2]][position]['score']+=s

        if (plus_strand >0) and (minus_strand ==0):
            alignment_count[i]['plus_aligned'] += dict_filt3_range_rounds[i][seq]
        if (plus_strand >0) and (minus_strand >0):
            alignment_count[i]['plus_and_minus'] += dict_filt3_range_rounds[i][seq]
        if (plus_strand ==0) and (minus_strand >0):
            alignment_count[i]['minus_aligned'] += dict_filt3_range_rounds[i][seq]
            
        if ((plus_strand == 0) and (minus_strand > 0)) or ((plus_strand > 0) and (minus_strand == 0)):
            for r2 in range(len(ribozyme)):
                if ribozyme_names[r2] in help_dict_align[seq]:
                    
                    my_range = alignment_end[seq][ribozyme_names[r2]] - alignment_start[seq][ribozyme_names[r2]]
                    for y in range(my_range):
                        position = alignment_start[seq][ribozyme_names[r2]]+y #getting position on ribozyme to put into dictionary for each nucleotide in sequence

                        s = dict_filt3_range_rounds[i][seq]/(count*len(help_dict_align[seq]))
                        dict_alignment_noplusandminus[i][ribozyme_names[r2]][position]['score']+=s
            
            
#turning dictionary of alignment to dataframe with mirrored complementary
import pandas as pd
rows = []

r_order = ['1_type5', '3_type5_reversecomp', '2_type1', '4_type1_reversecomp']

for pool in dict_alignment:
    for r in range(len(dict_alignment[pool])):
        for position in dict_alignment[pool][ribozyme_names[r]]:
            data_row = {}
            data_row['ribozyme']= r_order[r] 
            data_row['pool']=pool

            if data_row['ribozyme']=='4_type1_reversecomp':
                mirror_position = len(type1)-position-1
                data_row['position']= mirror_position
            elif data_row['ribozyme'] =='2_type5_reversecomp':
                mirror_position = len(type5)-position-1
                data_row['position']= mirror_position
            else:
                data_row['position']= position

            data_row['sequence'] = dict_alignment[pool][ribozyme_names[r]][position]['sequence']
            data_row['score']=dict_alignment[pool][ribozyme_names[r]][position]['score']

            rows.append(data_row)

df = pd.DataFrame(rows) 
df['position']=df['position']+1 #makes numbers more intuitive -> position 1 is 1 and not 0

df['position_seq']=df['position'].astype(str)+'_'+df['sequence'].astype(str)
df.to_csv('Ribozyme_align.csv')



# saving dataframe of alignment count
df = pd.DataFrame(alignment_count)

df.index.name ='type'
df.reset_index(inplace=True)
df_long = pd.melt(df,id_vars=['type'],var_name='pool', value_name='value')


df_long.to_csv("Alignment_count_all.csv")
```

```python
all_dict = {}
all_dict['1-ribo']={}
for pool in dict_ribozyme:
    all_dict['1-ribo'][pool]={}
    for ribo in dict_ribozyme[pool]:
        for seq in dict_ribozyme[pool][ribo]:
            all_dict['1-ribo'][pool][seq]=dict_ribozyme[pool][ribo][seq]
all_dict['2-grey']=dict_grey_area
all_dict['3-non_ribo']= dict_not_ribozyme

#complementary sequences with more than 8 nucleotides
import sys
from Bio.Seq import Seq
count_comp = {}

for split in all_dict:
    count_comp[split]={}
    for i in all_dict[split]:
        count_comp[split][i]={}
        count_comp[split][i]['unique_complementary'] = 0
        count_comp[split][i]['total_unique_seq']=0
        count_comp[split][i]['complementary_with_repeat'] = 0
        count_comp[split][i]['total_number_seq']=0
        for seq in all_dict[split][i]:

            if len(seq)>8:
                count_comp[split][i]['total_unique_seq']+=1
                count_comp[split][i]['total_number_seq'] +=dict_filt3_range_rounds[i][seq]
                my_DNA = Seq(seq)
                rev_complement = my_DNA.reverse_complement()
                if rev_complement in dict_filt3_range_rounds[i]:
                    count_comp[split][i]['unique_complementary']+=1
                    count_comp[split][i]['complementary_with_repeat']+=dict_filt3_range_rounds[i][seq]


        count_comp[split][i]['fraction_unique_complementary_seq']= count_comp[split][i]['unique_complementary']/count_comp[split][i]['total_unique_seq']
        count_comp[split][i]['fraction_complementary_seq_with_repeat']= count_comp[split][i]['complementary_with_repeat']/count_comp[split][i]['total_number_seq']

import pandas as pd
rows = []

for split in count_comp:
    for pool in count_comp[split]:
        for cat in count_comp[split][pool]:
            
            data_row = {}
            data_row['split']= split
            data_row['pool']=pool
            data_row['category']=cat
            data_row['value']=count_comp[split][pool][cat]
            rows.append(data_row)

df = pd.DataFrame(rows) 

df.to_csv('Complementarity_counts_splits.csv')
```
